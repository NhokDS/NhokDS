## Question 5: Are Balance Changes Logically Consistent?

```python
# Calculate the expected change in initiator's balance
df['initiator_balance_change'] = df['oldBalInitiator'] - df['newBalInitiator']

# Calculate the expected change in recipient's balance
df['recipient_balance_change'] = df['newBalRecipient'] - df['oldBalRecipient']

# Compare the calculated initiator balance change with the transaction amount
df['initiator_discrepancy'] = df['initiator_balance_change'] - df['amount']

# Compare the calculated recipient balance change with the transaction amount
df['recipient_discrepancy'] = df['recipient_balance_change'] - df['amount']

# Define a small tolerance for floating-point comparisons
tolerance = 1e-9

# Analyze initiator discrepancies
initiator_discrepancy_count = (abs(df['initiator_discrepancy']) > tolerance).sum()
print(f"Number of transactions where initiator's balance change does not match the amount (excluding small floating-point errors): {initiator_discrepancy_count}")

if initiator_discrepancy_count > 0:
    print("\nExamples of initiator discrepancies:")
    display(df[abs(df['initiator_discrepancy']) > tolerance].head())

    print("\nInitiator discrepancies by transaction type:")
    initiator_discrepancies_by_type = df[abs(df['initiator_discrepancy']) > tolerance]['transactionType'].value_counts()
    display(initiator_discrepancies_by_type)


# Analyze recipient discrepancies
recipient_discrepancy_count = (abs(df['recipient_discrepancy']) > tolerance).sum()
print(f"\nNumber of transactions where recipient's balance change does not match the amount (excluding small floating-point errors): {recipient_discrepancy_count}")

if recipient_discrepancy_count > 0:
    print("\nExamples of recipient discrepancies:")
    display(df[abs(df['recipient_discrepancy']) > tolerance].head())

    print("\nRecipient discrepancies by transaction type:")
    recipient_discrepancies_by_type = df[abs(df['recipient_discrepancy']) > tolerance]['transactionType'].value_counts()
    display(recipient_discrepancies_by_type)


# For 'TRANSFER' and 'PAYMENT' transaction types, we expect initiator's balance to decrease by 'amount'.
# For 'DEPOSIT' and 'WITHDRAWAL', the initiator balance behavior might be different.

# Let's specifically check consistency for TRANSFER/PAYMENT for initiator
print("\nConsistency check for TRANSFER/PAYMENT (initiator):")
transfer_payment_initiator_issues = df[
    ((df['transactionType'] == 'TRANSFER') | (df['transactionType'] == 'PAYMENT')) &
    (abs(df['initiator_discrepancy']) > tolerance)
]
print(f"Number of TRANSFER/PAYMENT transactions with initiator balance issues: {len(transfer_payment_initiator_issues)}")
if not transfer_payment_initiator_issues.empty:
    display(transfer_payment_initiator_issues['transactionType'].value_counts())

# Let's specifically check consistency for DEPOSIT/WITHDRAWAL for recipient (where recipient balance should increase/decrease by amount)
print("\nConsistency check for RECIPIENT (recipient): TRANSFER/DEPOSIT where recipient receives amount")
transfer_deposit_recipient_issues = df[
    ((df['transactionType'] == 'TRANSFER') | (df['transactionType'] == 'DEPOSIT')) &
    (abs(df['recipient_discrepancy']) > tolerance)
]
print(f"Number of TRANSFER/DEPOSIT transactions with recipient balance issues: {len(transfer_deposit_recipient_issues)}")
if not transfer_deposit_recipient_issues.empty:
    display(transfer_deposit_recipient_issues['transactionType'].value_counts())



# Overall statistics of discrepancies
print("\nStatistics of initiator balance-amount discrepancy:")
display(df['initiator_discrepancy'].describe())

print("\nStatistics of recipient balance-amount discrepancy:")
display(df['recipient_discrepancy'].describe())
```

## Comparison of 'amount' statistics for fraudulent and non-fraudulent transactions

```python
# Separate data into fraudulent and non-fraudulent transactions
df_fraud = df[df['isFraud'] == 1]
df_non_fraud = df[df['isFraud'] == 0]

# Define a function to calculate extended summary statistics (including percentiles and IQR)
def get_extended_summary(data, column_name):
    summary = data[column_name].agg(['mean', 'median', 'std', 'min', 'max'])
    percentiles_25 = data[column_name].quantile(0.25)
    percentiles_75 = data[column_name].quantile(0.75)
    iqr = percentiles_75 - percentiles_25

    summary['25th Percentile'] = percentiles_25
    summary['75th Percentile'] = percentiles_75
    summary['IQR'] = iqr
    return summary

# Get summary statistics for fraudulent transactions
print("\nSummary Statistics for Fraudulent Transactions (amount):")
fraud_summary = get_extended_summary(df_fraud, 'amount')
display(fraud_summary)

# Get summary statistics for non-fraudulent transactions
print("\nSummary Statistics for Non-Fraudulent Transactions (amount):")
non_fraud_summary = get_extended_summary(df_non_fraud, 'amount')
display(non_fraud_summary)

# Combine for easier comparison
comparison_df = pd.DataFrame({
    'Fraudulent': fraud_summary,
    'Non-Fraudulent': non_fraud_summary
})
print("\nComparison of 'amount' statistics:")
display(comparison_df)
```

### Visualizing Comparison of 'amount' Statistics

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Transpose the comparison_df for easier plotting
comparison_df_T = comparison_df.T

# Create subplots for each statistic
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 15))
axes = axes.flatten() # Flatten the 2x2 array of axes for easy iteration

metrics_to_plot = ['mean', 'median', 'std', 'min', 'max', '25th Percentile', '75th Percentile', 'IQR']

for i, metric in enumerate(metrics_to_plot):
    sns.barplot(x=comparison_df_T.index, y=comparison_df_T[metric], ax=axes[i], palette='viridis')
    axes[i].set_title(f'{metric.replace("_", " ").title()} of Amount')
    axes[i].set_ylabel('Amount')
    axes[i].set_xlabel('')
    # Add value labels on top of bars
    for container in axes[i].containers:
        axes[i].bar_label(container, fmt='%.2f', label_type='edge')

# Remove any unused subplots
for j in range(len(metrics_to_plot), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle('Comparison of Transaction Amount Statistics: Fraudulent vs. Non-Fraudulent', y=1.02, fontsize=16)
plt.show()
```

### 7. Discussion: Variables Useful for Fraud Detection and Potentially Misleading Variables

Based on the comprehensive visual and statistical exploration of relationships between variables, we can identify several key factors highly indicative of fraudulent activity and some that require careful interpretation.

#### **Variables Potentially Useful for Fraud Detection:**

1.  **`transactionType`:**
    *   **Insight:** The `fraud_rate_by_type` analysis clearly showed that **'TRANSFER' transactions are overwhelmingly the most susceptible to fraud** (approx. 45.82% fraud rate), while other types had negligible to zero fraud. The boxplots and separate histograms for `isFraud` confirmed that fraudulent amounts tend to cluster in a higher range, with the median fraudulent amount being substantially greater than the median non-fraudulent amount.
    *   **Usefulness:** A crucial numerical feature. Its raw value, log-transformed value (`log_amount`), or binned versions could be highly predictive. Thresholds based on `amount` (especially in conjunction with `transactionType`) could be very effective.

3.  **`oldBalInitiator` and `newBalInitiator` (in relation to `amount`):**
    *   **Insight:** The scatter plots revealed that many fraudulent transactions occur when the initiator's `oldBalInitiator` or `newBalInitiator` is very low (e.g., zero or even negative).     *   **Usefulness:** The *relationship* between `amount` and initiator balances is highly informative. Features like `initiator_discrepancy` (calculated earlier) explicitly quantify inconsistencies in initiator balances relative to the transaction amount, which are strong fraud signals.

1.  **`initiator` and `recipient` (IDs):**
    *   **Insight:** These are unique identifiers. While useful for linking transactions or tracking entities, their raw values have no direct predictive power. Using them directly in models might lead to overfitting if not properly handled (e.g., through one-hot encoding for categorical models, or feature engineering to derive behavioral features per user).
    *   **Misleading Aspect:** Simply including them as high-cardinality categorical features without proper handling (e.g., target encoding, or using them to derive features like 'number of transactions by this initiator' or 'average amount for this recipient') could confuse the model or consume excessive memory/computation.

2.  **`amount` (raw value for non-fraudulent transactions, or without log transformation):**
    *   **Insight:** The distribution of `amount` is highly skewed, especially for non-fraudulent transactions. The mean is significantly pulled up by a few large legitimate transactions. The boxplot of `amount` by `isFraud` also shows that non-fraudulent transactions have extremely high outliers that overlap with fraudulent amounts.
    *   **Misleading Aspect:** Relying solely on the mean or using raw `amount` in models sensitive to feature scale (like linear models or neural networks) without transformation could give disproportionate weight to rare, large legitimate transactions, potentially increasing false positives. Similarly, a simple threshold on raw `amount` might miss lower-value fraud or flag legitimate large transactions.

3.  **`oldBalInitiator` and `newBalInitiator` (without considering `amount` and `transactionType`):**
    *   **Insight:** These balances themselves can be very large or negative for legitimate reasons (e.g., a large legitimate account, or an account with a temporary overdraft). Simply flagging all negative balances might lead to too many false positives without context.
    *   **Misleading Aspect:** Interpreting these balances in isolation can be misleading. Their significance for fraud largely comes from their *relationship* to the transaction `amount` and the `transactionType` (e.g., a large `amount` being transferred from a zero or negative `oldBalInitiator` for a `TRANSFER` transaction is suspicious; a negative balance for a `PAYMENT` might be less so). The `initiator_discrepancy` feature is much more informative than the raw balances alone.

In summary, features related to `transactionType`, `amount`, and balance consistency (`initiator_discrepancy`, `recipient_discrepancy`) appear to be the most promising for fraud detection. Temporal features derived from `step` could offer supplementary insights. Raw identifiers and highly skewed numerical features need careful engineering and interpretation to avoid misleading the detection model.

2.  **Negative Initiator Balances:** A considerable percentage of `oldBalInitiator` (~27.8%) and `newBalInitiator` (~31.5%) transactions have negative balances. ### Why Visual Exploration is Necessary Even When Summary Statistics are Available:

Visual exploration is indispensable because:

1.  **Reveals Hidden Patterns:** Summary statistics (mean, median, std) can mask complex distributions (e.g., multimodal, highly skewed) and non-linear relationships that are immediately apparent through plots like histograms, scatter plots, and boxplots.
2.  **Identifies Outliers and Anomalies:** Visualizations clearly highlight outliers, data entry errors, or unusual data points that might skew statistical measures or indicate data quality issues, but are not explicitly captured by simple statistics.
3.  **Communicates Insights Effectively:** Graphs and charts provide an intuitive and powerful way to communicate complex data characteristics and findings to both technical and non-technical stakeholders, fostering a deeper understanding of the data.

5.  **Confirms Assumptions:** Visuals help confirm or challenge assumptions about data distributions and relationships, which is crucial for selecting appropriate statistical methods or machine learning models.
